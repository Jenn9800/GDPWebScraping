```python
from bs4 import BeautifulSoup
import requests
```

```python
url ='https://www.geeksforgeeks.org/worlds-top-richest-countries/'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
print(soup)
```

```python
soup.find('table', class_='GFGEditorTheme__table')
#<table class="GFGEditorTheme__table">
```

```python
table = soup.find_all('table')[0]
```

```python
world_titles = table.find_all('th')
world_titles
```

```python
world_table_titles = [title.text.strip() for title in world_titles] #strip: list to string; split: list to list
print(world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:  #start from 1 not from 0
    row_data = row.find_all('td')
    ind_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = ind_row_data

df
```

```python
df.to_csv('GDPWebScraping.csv', index=False)
```
